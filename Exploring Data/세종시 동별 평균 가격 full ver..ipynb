```python
import pandas as pd
df1 = pd.read_csv('./complex_20230308.csv', header=None)
df1.columns = ['old_addr_idx', 'complex_name', 'property_type','시도','시군구',
               '읍면동','리','road_name','road_code','admin_dong_code','postal_code','jibun_main','jibun_sub','updated']

df_sejong = df1.loc[df1['시도']=='세종특별자치시']
```

```python
df2 = pd.read_csv('./property_20230308.csv', header=None)
df2.columns = ['pk','old_addr_idx','dong','ho','official_price','평수','updated']

df2
```

```python
df_sejong_dong = df2.merge(df_sejong, on='old_addr_idx')
df_sejong_dong
```

```python
df_sejong_dong_apt = df_sejong_dong[df_sejong_dong['property_type'] == 'apartment']
df_sejong_dong_apt
```

```python
df_sejong_dong_apt_avg = df_sejong_dong_apt.groupby(['읍면동']).mean()
df_sejong_dong_apt_avg
```

```python
df_sejong_dong_apt_avg.iloc[0]
```

```python
null_counts = df_sejong_dong_apt.isnull().sum()
null_counts
```

```python
df_sejong_dong_apt_avg = df_sejong_dong_apt_avg.drop(columns='평수', axis=1)

df_sejong_dong_apt_avg_plt = df_sejong_dong_apt_avg.reset_index()
df_sejong_dong_apt_avg_plt_sort = df_sejong_dong_apt_avg_plt.sort_values('official_price', ascending=False)
df_sejong_dong_apt_avg_plt_sort
```

```python
pd.set_option('display.float_format', lambda x: f'{x:,.0f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
graph = df_sejong_dong_apt_avg_plt_sort.plot(kind='bar', x='읍면동', y='official_price', width = 0.5, legend = None, figsize=(10,6))
plt.rc('font', family="Malgun Gothic") 
plt.title('세종시 동별 아파트 평균 가격')
plt.ylabel('가격 (단위:백만원)')
graph.xaxis.set_tick_params(rotation=45, labelsize=10)


# Define a function to format the tick labels
def millions(x, pos):
    'The two args are the value and tick position'
    return '{:.1f}M'.format(x*1e-6)

# Format the tick labels on the y-axis
graph.yaxis.set_major_formatter(ticker.FuncFormatter(millions))
```
